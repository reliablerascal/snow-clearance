# 1. Read and Review all Datasets

```python
import pandas as pd
```

```python
df_addresses = pd.read_csv("../results/fines_by_address.csv")
df_dockets = pd.read_csv("../results/fines_by_docket.csv")
# df_311_snow = pd.read_csv("../data/311-complaints-snow.csv")
df_311_vacant = pd.read_csv("../data/311-complaints-vacant.csv")
#df_vacant_violations = pd.read_csv("../data/vacant-violations.csv")
```

### filter by Englewood

```python
df_englewood = df_dockets[(df_dockets['community']=='ENGLEWOOD')]
```

### fines by department

```python
df_englewood.groupby("dept").size().reset_index(name="n")
```

### filter by streets and sanitation

```python
df_englewood_ss = df_englewood[(df_englewood['dept']=='STRTSAN')]
```

```python
len(df_englewood_ss)
```

```python
df_englewood_ss.head()
```

### 311 vacant property complaints

```python
df_311_vacant.head()
```

# 2. Match Dockets to 311 complaints about vacants

```python
df_match_englewood = pd.merge(df_englewood_ss,df_311_vacant,left_on="address",right_on="STREET_ADDRESS", how="left")
df_match_englewood
```

```python
df_match_englewood.to_csv("../results/englewood/fines_streets_englewood.csv", index=False)
```
