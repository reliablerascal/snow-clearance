```python
df = pd.read_csv("../data/311-snow-violations.csv")
df.head()
```

### look at just full years

```python
df = df[(df['season'].isin(['2019-2020','2020-2021','2021-2022','2022-2023']))]
len(df)
```

# 2. By Ward, 2019-2023

```python
df.groupby('WARD').size()
```

```python
df_ward_by_season= df.pivot_table(index='WARD',columns='season',values='SR_NUMBER',aggfunc='count').reset_index()
df_ward_by_season
```

```python
df_ward_by_season.to_csv("../results/ward_by_season.csv", index=False)
```
