My goal here is to summarize ward-level data on CDOT fines in one place, to inform outreach to alders. I'm looking for strictly numbers here. I had done this previously with community areas

1) roll up to unique dockets
2) merge with snow complaints
3) calculate complaint to docket ratio
4) group by ward

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/03-geocoded/fines-joined-to-wards-2015.csv")
df
```

```python
df ['Docket Number'].nunique()
```

```python
df_dockets = df.groupby(['Docket Number','season','ward','Violation Date','Violation Address','Issuing Department Code']).size().reset_index()
df_dockets = df_dockets.rename(columns={'Docket Number':'docket','Issuing Department Code':'dept'})
df_dockets['ward'] = df_dockets['ward'].astype(int)
df_dockets
```

```python
df_ward_dockets = df_dockets.pivot_table(index='ward',
                             columns='dept',
                             values='docket',
                             aggfunc='count',
                             fill_value=0).reset_index()
df_ward_dockets.columns.name = None
df_ward_dockets['total_dockets']=df_ward_dockets['BAFCONP']+df_ward_dockets['POLICE']+df_ward_dockets['STRTSAN']+df_ward_dockets['TRANPORT']
df_ward_dockets.head()
```
