# 1. Read Dockets Detail Dataset

```python
import pandas as pd
```

```python
df_dockets2= pd.read_csv('../data/04-standardized/dockets-detailed.csv')
```

# 2. Summarize Respondents to look for repeat offenders

```python
# create unduplicated dockets to respondents recordset
df_respondents = df_dockets2[['docket','respondent']].drop_duplicates()
df_respondents.head()
```

```python
len(df_respondents)
```

```python
# find dockets with the most # of respondents
df_respondents.groupby('docket').size().reset_index(name='n').sort_values(by='n',ascending=False).head()
```

```python
# what does docket 22DS03847M look like
df_respondents[df_respondents['docket']=='22DS03847M'].reset_index()
```

```python
df_respondents_count = df_respondents.groupby('respondent').agg(n=('docket','count')).reset_index().sort_values('n', ascending = False)
```

```python
df_respondents_count
```

# 3. export useful datasets for further analysis

```python
df_respondents_count.to_csv("../results/respondents-count.csv", index = False)
```
