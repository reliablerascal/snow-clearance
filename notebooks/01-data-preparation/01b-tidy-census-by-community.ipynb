### keep only worthwhile fields

```python
df_census = df[['GEOG','2020_POP','VAC_HU','HU_TOT','VACANT','VACperc']]
df_census['community_caps']=df_census['GEOG'].str.upper()
df_census.columns = df_census.columns.str.lower()
df_census = df_census.rename(columns={'geog':'community_name'})
df_census
```

### Replace 'The Loop' with 'Loop', to match community names in other data sets

```python
# boundaries shapefile refers to downtown as "LOOP", not "The Loop"
df_census['community_name']= df_census['community_name'].replace('The Loop','Loop')
df_census['community_caps']= df_census['community_caps'].replace('THE LOOP','LOOP')
df_census
```
