# Replace Department Names on Map
3/24/24

```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/05-finalized/dockets-summary.csv")
```

```python
df.head()
```

```python
def dept_label(df):
    replacements = {
        'TRANPORT': 'CDOT',
        'STRTSAN': 'DSS',
        'POLICE': 'Police',
        'BAFCONP': 'BACP',
    }
    df['dept_label'] = df['dept'].replace(replacements)
    return df
```

```python
df = dept_label(df)
```

```python
df.head()
```

```python
#drop the one unknown record, for simplicity
df= df[df['dept']!='unknown']
```

```python
df.to_csv("../../results/ssw02-fines/citations-map.csv",index=False)
```
